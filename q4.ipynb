```python
def string_reverse(s):
    """
    Task 1
    - Create a function that reverses a given string (s).
    - s must be a string.
    - Return the reversed string.
    """

    # taking reference from https://www.w3schools.com/python/python_howto_reverse_string.asp
    # [::-1] where -1 means step size of 1 in backward direction
    reversed_s = s[::-1]
    return reversed_s


# Task 2
# Invoke the function "string_reverse" using the following scenarios:
# - "Hello World"
# - "Python"
print(string_reverse("Hello World"))
print(string_reverse("Python"))
```
