```python
def update_dictionary(dct, key, value):
    """
    Task 1
    - Create a function that updates a dictionary (dct) with a new key-value pair.
    - If the key already exists in dct, print the original value, then update its value.
    - Return the updated dictionary.
    """
    for key in dct.keys():                              # if key already exists
      print(f'original {key}: ' + str(dct.get(key)))    # print original value, convert to str before it can be concatenated for print

    dct.update({key: value})      # update value corresponding to the key

    return dct


# Task 2
# Invoke the function "update_dictionary" using the following scenarios:
# - {}, "name", "Alice"
# - {"age": 25}, "age", 26
print(update_dictionary({}, "name", "Alice"))
print(update_dictionary({"age": 25}, "age", 26))
```
